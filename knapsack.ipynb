```python
data_filtered = data[['item', 'calories', 'protien', 'menu']]
data_filtered
```

```python
def knapsack_greedy(items, weight_column, profit_column, capacity):
    items = items.sort_values(by=[profit_column, weight_column], ascending=[False, True])
    
    total_weight = 0
    total_profit = 0
    selected_items = []
    
    for _, item in items.iterrows():
        if total_weight + item[weight_column] <= capacity:
            total_weight += item[weight_column]
            total_profit += item[profit_column]
            selected_items.append(item['item'])
    
    return selected_items, total_weight, total_profit
    
```

```python

def index():
    if request.method == "POST":
        # Get user input for capacity
        capacity = int(request.form["capacity"])
        menu = request.form["menu"]
        
        # Filter data by selected menu
        menu_data = data_filtered[data_filtered['menu'] == menu]
        
        # Apply knapsack algorithm
        selected_items, total_weight, total_profit = knapsack_greedy(menu_data, 'calories', 'protien', capacity)
        
        # Return results to frontend
        return render_template("index.html", selected_items=selected_items, total_weight=total_weight, total_profit=total_profit)
    
    return render_template("index.html", selected_items=None)
```
